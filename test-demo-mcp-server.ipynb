```python
import yaml
from fastmcp import Client

client = Client("http://127.0.0.1:8085/mcp")

async with client as c:

    demos = await c.call_tool("list_demos")
    yaml_text = yaml.safe_dump(demos.structured_content or demos.data, sort_keys=False)
    print("Demos:", yaml_text)

    demo = await c.call_tool(
        "get_demo_details",
        {"key": "MCP Server Client Orchestrator", "environment": "local"}
    )

    yaml_text = yaml.safe_dump(demo.structured_content or demo.data, sort_keys=False)
    print("get_demo_details:", yaml_text)

    demo = await c.call_tool(
        "get_demo_prerequisites",
        {"key": "MCP Server Client Orchestrator", "environment": "local"}
    )

    yaml_text = yaml.safe_dump(demo.structured_content or demo.data, sort_keys=False)
    print("get_demo_prerequisites:", yaml_text)
```
